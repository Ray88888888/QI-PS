```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
import datetime
import pingouin as pg
```

## Part I (Ray & Zhewei)

```python
df25=pd.read_csv('https://raw.githubusercontent.com/Ray88888888/QI-PS/main/25_Portfolios_5x5.CSV', header=0,index_col=0)
df25.replace([-99.99,-999],[np.nan,np.nan],inplace=True)
df25 = df25[:202208]
df25
```

```python
df25_mean = []
for i in range(25):
  df25_mean.append(df25[str(i+1)].mean())
df25_mean
```

```python
'''Report mean returns'''
def present_in_5square(df25):
  df25_return = pd.DataFrame(index=['Small','2','3','4','Big'], columns=['Low','2','3','4','High'], dtype=float)
  for i in range(25):
    n = i//5
    m = i%5
    df25_return.iloc[n][m] = df25[i]
  return df25_return

present_in_5square(df25_mean)
```

```python
name_of_25_port =[]
for i in ['Small','2','3','4','Big']:
  for j in ['Low','2','3','4','High']:
    name_of_25_port.append(i+'_'+j)
name_of_25_port
```

```python
'''Run OLS time-series regressions to find parameter estimates of αˆ,βˆ,λˆ'''
'''test CAPM & Report standard errors'''
df_timeseries_est = pd.DataFrame(index=df25.columns[:-4], columns=['α','α_pvalue','β','β_pvalue','α_se','β_se'], dtype=float)
for i in df25.columns[:-4]:
  lm = pg.linear_regression(df25['Mkt-RF'],df25[i]-df25['RF'])
  df_timeseries_est['α'][i] = lm.coef[0]
  df_timeseries_est['α_pvalue'][i] = lm.pval[0]
  df_timeseries_est['β'][i] = lm.coef[1]
  df_timeseries_est['β_pvalue'][i] = lm.pval[1]
  df_timeseries_est['α_se'][i] = lm.se[0]
  df_timeseries_est['β_se'][i] = lm.se[1]
for i in df_timeseries_est.columns:
  print(i)
  print(present_in_5square(df_timeseries_est[i]).round(4))
  print('\n')

print('λ')
print(df25['Mkt-RF'].mean())
print('λ_se')
print(df25['Mkt-RF'].std()/(len(df25.index)**0.5))
```

```python
'''test FF3F & Report standard errors'''
df_timeseries_est = pd.DataFrame(index=df25.columns[:-4], columns=['α','α_pvalue','β','β_pvalue','s','s_pvalue','h','h_pvalue','α_se','β_se','s_se','h_se'], dtype=float)
for i in df25.columns[:-4]:
  lm = pg.linear_regression(df25[['Mkt-RF','SMB','HML']],df25[i]-df25['RF'])
  df_timeseries_est['α'][i] = lm.coef[0]
  df_timeseries_est['α_pvalue'][i] = lm.pval[0]
  df_timeseries_est['β'][i] = lm.coef[1]
  df_timeseries_est['β_pvalue'][i] = lm.pval[1]
  df_timeseries_est['s'][i] = lm.coef[2]
  df_timeseries_est['s_pvalue'][i] = lm.pval[2]
  df_timeseries_est['h'][i] = lm.coef[3]
  df_timeseries_est['h_pvalue'][i] = lm.pval[3]
  df_timeseries_est['α_se'][i] = lm.se[0]
  df_timeseries_est['β_se'][i] = lm.se[1]
  df_timeseries_est['s_se'][i] = lm.se[2]
  df_timeseries_est['h_se'][i] = lm.se[3]

for i in df_timeseries_est.columns:
  print(i)
  print(present_in_5square(df_timeseries_est[i].round(4)))
  print('\n')

print('λ for market')
print(df25['Mkt-RF'].mean())
print('λ_se')
print(df25['Mkt-RF'].std()/(len(df25.index)**0.5))
print('\n')

print('λ for size')
print(df25['SMB'].mean())
print('λ_se')
print(df25['SMB'].std()/(len(df25.index)**0.5))
print('\n')

print('λ for value')
print(df25['HML'].mean())
print('λ_se')
print(df25['HML'].std()/(len(df25.index)**0.5))
print('\n')
```

```python
df_beta = []
for i in df25.columns[:-4]:
  lm = pg.linear_regression(df25['Mkt-RF'],df25[i]-df25['RF'])
  df_beta.append(lm.coef[1])

df_alpha = []
for i in df25.columns[:-4]:
  lm = pg.linear_regression(df25['Mkt-RF'],df25[i]-df25['RF'])
  df_alpha.append(lm.coef[0])
```

```python
'''Cross sectional test with a constant'''
lm = pg.linear_regression(df_beta,df25_mean)
lm['names'] = ['λ0^','λ^']
lm
```

```python
alpha = []
for i in range(25):
  alpha.append(df25_mean[i] - df_beta[i]*lm.coef[1] - lm.coef[0])

'''mean square'''
mean_square = 0
for i in alpha:
  mean_square += (i-sum(alpha)/len(alpha))**2
print('mean square')
print(mean_square/len(alpha))
print('\n')

'''average absolute'''
average_absolute = 0
for i in alpha:
  if i-(sum(alpha)/len(alpha)) >= 0:
    average_absolute += i-(sum(alpha)/len(alpha))
  else:
    average_absolute += (sum(alpha)/len(alpha)) - i
print('average absolute')
print(average_absolute/len(alpha))
```
